# Intro to matplotlib

```python
import math
import random
import matplotlib.pyplot as plt
```

```python
tsla = 657.62
sigma = .30
T = 1 / 252
r = 0.00125
prices = []
```

```python

for day in range(63):
    prices.append(tsla)
    tsla *= math.exp((r - .5 * sigma ** 2)* T + sigma * math.sqrt(T)* random.normalvariate(0,1))
   
```

```python
prices
```

```python
# basic line plot
plt.plot(prices)
```

```python
# basic histogram
data = [random.normalvariate(0,1) for x in range(100000)]
```

```python
plt.hist(data, bins = 25, color = 'g', edgecolor = 'w', alpha = .7);
```

```python
x = [100 + random.normalvariate(0,1) for point in range(60)]
y = [point + random.normalvariate(0,1) for point in x]
```
